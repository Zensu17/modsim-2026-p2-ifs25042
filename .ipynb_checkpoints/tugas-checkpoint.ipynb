```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_excel("data_kuesioner.xlsx")

# Pisahkan kolom pertanyaan (asumsi kolom pertama adalah ID/Timestamp)
questions = df.columns[1:] 
data_questions = df[questions]

# Mapping untuk kategori (Sesuaikan dengan skala kuesioner Anda)
kategori_map = {1: 'Negatif', 2: 'Negatif', 3: 'Netral', 4: 'Positif', 5: 'Positif'}
```

```python
# Flatten data untuk melihat distribusi total
total_counts = data_questions.stack().value_counts().sort_index()

fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Bar Chart
total_counts.plot(kind='bar', ax=ax[0], color='skyblue', edgecolor='black')
ax[0].set_title('Distribusi Jawaban (Keseluruhan)')
ax[0].set_xlabel('Skor')
ax[0].set_ylabel('Frekuensi')

# Pie Chart
total_counts.plot(kind='pie', ax=ax[1], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
ax[1].set_title('Proporsi Jawaban (Keseluruhan)')
ax[1].set_ylabel('')

plt.show()
```

```python
categorized_data = data_questions.map(lambda x: kategori_map.get(x, 'Lainnya'))
cat_counts = categorized_data.stack().value_counts().reindex(['Positif', 'Netral', 'Negatif']).fillna(0)
```

```python
# Stacked Bar: Distribusi per Pertanyaan
dist_per_q = data_questions.apply(lambda x: x.value_counts()).T.fillna(0)
dist_per_q.plot(kind='barh', stacked=True, figsize=(10, 7), colormap='viridis')
plt.title('Distribusi Jawaban per Pertanyaan')
plt.xlabel('Jumlah Responden')
plt.legend(title='Skor', bbox_to_anchor=(1.05, 1))
plt.show()

# Bar Chart: Rata-rata Skor per Pertanyaan
avg_scores = data_questions.mean().sort_values()
plt.figure(figsize=(10, 5))
avg_scores.plot(kind='bar', color='salmon')
plt.title('Rata-rata Skor per Pertanyaan')
plt.ylabel('Skor Rata-rata')
plt.xticks(rotation=45)
plt.show()
```
